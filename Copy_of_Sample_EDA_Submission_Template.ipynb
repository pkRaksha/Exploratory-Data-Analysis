## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

#Cleaning the data:

#Replacing all the undefined text in the dataframe with meaningful text
df.replace(to_replace="Undefined",value="Not provided",inplace=True)

# Iterating through the columns with missing and nan values 
for i in df.columns[df.isna().any()].tolist():
  if df[i].dtype!=object:                               # If its of numeric type replacing with zero
     df[i].fillna(0,inplace=True)
  else:                                                  # If not replacing with a text 
    df[i].fillna("Not Provided",inplace=True)


# Dropping all the duplicate rows 
new_df = df.drop_duplicates()

# Adding a new column to store the total  no of days stayed 
new_df['length_of_stay']=new_df['stays_in_weekend_nights']+new_df['stays_in_week_nights']

#Adding a column to store the total number of occupant in each booking
new_df['total_occupant']=new_df['adults']+new_df['babies']+new_df['children']

# Adding a new column date booked by combining the year and  month 
new_df['date_arrived']=new_df['arrival_date_year'].astype(str) + '-' + new_df['arrival_date_month']
````

#### Chart - 12

````python
# Chart - 12 visualization code

# Multi variate analysis

# hotel wise Year wise month wise ratio of cancellation and booking made

# No. of booked and cancelled count for each date is calculated individually
Table1=new_df.groupby(['hotel','date_arrived'])['is_canceled'].value_counts().reset_index(name='Num_canceled_and_booked')

# Total no of bookings made is calculated 
Table2=Table1.groupby(['hotel','date_arrived'])['Num_canceled_and_booked'].sum().reset_index(name='Total_booked')

# Merging the two tables 
Cancelation_info = pd.merge(Table1, Table2, how='outer', on=['hotel','date_arrived'])

# Calculating the cancellation rate 
Cancelation_info['%rate']=(Cancelation_info['Num_canceled_and_booked']*100/Cancelation_info['Total_booked']).round(decimals=2)

# Using the cancellation info table to calculate the cancellation rate over a period of time 
cancellation_rates=Cancelation_info.loc[Cancelation_info['is_canceled']==1,['%rate','hotel','date_arrived']]

#Plotting the cancellation rates
fig, ax = plt.subplots(figsize=(20, 5))
chart2=sns.lineplot(data=cancellation_rates,x='date_arrived',y='%rate',ax=ax,hue="hotel",style="hotel", markers=True, dashes=False)
chart2.set_xticklabels(labels=cancellation_rates['date_arrived'],rotation=30)
plt.title("Cancellation Rates over a the years 2015 - 2017")
plt.show()
````

To show the cancellation rtes over a period of time the line chart is best
